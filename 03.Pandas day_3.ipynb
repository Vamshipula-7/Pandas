**Merge Operations**

```python
#merge based on value,coloummn name same or diifferent it wll work
```

```python
d1={"key1":[1,2,3,4,5,6],
    "key2":[7,8,9,10,11,"Vamshi"],
    "key3":[13,14,15,16,17,18],
    "key4":[19,20,21,22,23,24]
   }
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(d1)
```

```python
df1
```

```python
d2={"key1":[1,2,3,4,5,6],
    "key2":[7,8,9,10,11,"Vamshi"],
    "key3":[13,14,15,16,"Krishna",18],
    "key4":[19,20,21,"Pula",23,24]
   }
```

```python
df2=pd.DataFrame(d2,index=["a","b","c","d","e","f"])
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='inner',on="key4")
```

```python
#left Merge
pd.merge(df1,df2,how="left",on="key4")
```

```python
[[df1,df2]]
```

```python
#Right Merge
pd.merge(df1,df2,how="right",on="key4")
```

```python
#outer Merge
pd.merge(df1,df2,how="outer",on=["key1","key4"])
```

```python
[[df1,df2]]
```

```python
d10={"key11":[1,2,3,4,5,"sudh"],
    "key21":[7,8,9,10,11,12],
    "key31":[13,14,15,16,17,18],
    "key41":[19,20,21,22,23,24]
   }
```

```python
d20={"key1":[1,2,3,4,5,"sudh"],
    "key2":[7,8,"kumar",10,11,12],
    "key3":[13,14,15,16,"Krishna",18],
    "key4":[19,20,21,"Pula","Data scientist",24]
   }
```

```python
df10=pd.DataFrame(d10,index=[1,2,3,4,5,6])
```

```python
df20=pd.DataFrame(d20,index=[1,2,3,4,5,"sudh"])
```

```python
df10
```

```python
df20
```

```python
df10.join(df20,how="right")
```

```python
df10.join(df20,how="left")
```

```python
df10.join(df20,how="outer")
```

```python
df10.join(df20,how="inner")
```
